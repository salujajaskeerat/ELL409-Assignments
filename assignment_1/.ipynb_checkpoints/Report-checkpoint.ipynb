# Part 1-A assignment

Fitting polynomial of degree M to points (x,y)
The points are transformed to space $\phi(x)$ = $1 + x + x^2 +x^3 ...x^M$ and the linear regression is applied to $(\phi(x),y)$.

Error is calculated using <span class="mark">least square regression 
Optimization</span> can be done :
1. Using moore-penrose inverse formuale
2. Using gradient descent algorithm

## Polynomial Fit for first 20 points

The data has gaussian noise with 0 mean i.e $N(h(x)|t,\frac{1}{\beta})$ , where $\frac{1}{\beta}$ is variance of the noise distribution

```python
data = np.genfromtxt('./gaussian.csv',delimiter=',')
X ,Y = data[:20,0],data[:20,1]
plt.scatter(X,Y,color='red')
plt.legend(['Y'])
plt.ylabel('Label')
plt.xlabel('X')
plt.show()
```

## **Normalizing Data**

After generating the design matrix we will normalize the features vectors i.e $x,x^2,x^3....x^M$ .
The feature X $\in$ [-M,M] , then the features $X^2 \in$ [-M^2,M^2] , and so on for all other powers of X. 
Hence the loss function (least squared error) forms contours with eliptical shapes . 

Hence while optimizing error in the gradient descent algorithm , **an improvement in** $X^2$ **may cause big step in**$X$, thus instead of decresing error ,we may cross the minma point , and error may increase . This would cause a lot of oscillation of $w$ around the minima point.

An **unnormalized** data would causes:
1. **highly sensitive learning parameter**
2. **high oscillation of paramter** $w$ during convergence


To prevent this we use *normalization techniques like* **z-normal** and *min-max normalization*. This transforma the contour plots of loss function to **circular shapes** , hence the gradient descent converges better.

**Z-Normalization** : The feature vector is tranformed as  $$(X) -> \frac{X-\mu}{\sigma}$$
where $\mu :$ mean of the feature vector and $\sigma$ : variance

**Min-Max Normalization** : The feature vector is tranformed as  $$(X) -> \frac{X-min(X)}{max(x)-min(x)}$$
where $\mu :$ mean of the feature vector and $\sigma$ : variance
Thus whole data is X lies in [-1,1]

**Z-Normalizing feature vector** : {$x,x^2,... x^M$}is transformed as  $$x^i -> \frac{(x^i-mean(x^i)}{\sigma(x^i}$$.

**Running gradient descent** yeild us $w_0^* ,w_1^* ... w_M^*$ . Which optimizes error for our normalized data points X. Thus we need to '**un-normalize**' the paramters $W$ to previous space

The unnormalized paramter we get are $$w_i = \frac{w_i^*}{\sigma_i} , \forall i \in \{1,M\} and $$
$$w_0 = w_0^* - \sum_{i=1}^{i=M} {\frac{\mu_i}{\sigma_i}}$$

**Note** : <span class="mark">In my implementation of function</span> polyfit class the feature vector is *auto normalized* before training and parameters are **auto unnormalized** after the training , this gives stable and faster convergence

**Design matrix**

Below is normalized design matrix with $M=3$ using **z-normalization**, The design matrix has (apprx)0 mean as below 

```python
from normalize import *
deg=3
phi= (np.power(np.reshape(X,(-1,1)),np.arange(0,deg+1)))
phi_norm = norm_data(phi,method='znormal')
print("MEAN of feature vector",np.mean(phi_norm.phi[:,1:],axis=0),'\n')

print("Design matrix",phi)
```

## Error Minimization on given number of points


### (a) Optimization using **Batch Gradient descent** on given data

Notice that convergence of gradient descent algorithm is quite slow , took around 50,000 iterations on an average to converge.

```python
degrees = np.arange(0,20)
models,w_ini =[],np.zeros(1).T
for degree in degrees:
    model = polyfit(degree=degree,max_iter=50000,steps=1e-3)
    model.fit(X,Y,w_ini=w_ini)
    w_ini=np.append(model.w,[[0]],axis=0)
    models.append(model)
```

```python
#After training lets plot all the curves

plt.figure(figsize=(18,8))
#Plotting curves
plt.scatter(X,Y,label='Actual Distrib',color='red')
for i in  (0,4,8,12,16,19):
    plt.plot(np.sort(X),models[i].pred(np.sort(X)),label='degree ='+str (models[i].degree)+(" Train error = ")+str(np.float16(models[i].train_error)))
plt.legend(),plt.title('PolyFit on First 20 Points'),plt.xlabel('X'),plt.ylabel('Y')
plt.show()
```

```python
grad_error = [model.train_error for model in models]
plt.figure(figsize=(15,5))
plt.plot(np.arange(0,len(grad_error)),grad_error,'-o')
plt.title("Gradient Descent: Training error vs Degree"),plt.xlabel("Degree of polynomial"),plt.ylabel("E(W) training error")
plt.show()
```

#### Observations:
1. **Starting degree 8 onwards** the training error is reduced drastically , hence  $degree\geq 8 $ are good fit to our data set
2. Training error reduces as the degree of poynomial is increased ,but this occurs at the expense of the increase complexity of model.

### **Variation of error vs iteration  of gradient descent** 

From below we can say the gradient descent algorithm converges very fast for the first 5000 iteration , after that the convergence of gradient descent is very slow in successive iteration.
Here 1 iteration corresponds to 1 pass over the data set , hence in our case 1 iteration implies 1 pass over all 20 points.

Error is monotonically decresing as number of iterations increases


```python
start,gaps,max_iter=100,1000,70000
fig,axes =plt.subplots(nrows=2,ncols=2,figsize=(16,8))
axes=np.array(axes)

for degree ,ax in zip([6,10,14,19],axes.reshape(-1)):
    iters = np.arange(start,max_iter,gaps)
    modeli = polyfit(degree=degree,max_iter=max_iter,steps=5e-4)
    modeli.fit(X,Y,error_gaps=gaps)
    iter_errors = modeli.train_error_gapped[1:]
    ax.plot(iters,iter_errors,'-o',color='red')
    ax.set_xlabel('Iterations'),ax.set_ylabel('Least squared Error'),ax.legend(['Degree = '+str(degree)])
```

## Good fit of polynomial (Underfitting , overfitting) (without regularization)

We know the  $\textbf{train error} -> 0$ as the degree of poynomial M increases but this lead to overfitting of the model over the underlying data set.

**Test validation** 
To test a model we will partition our data i.e 20 points into test and train data . The model is trained on trained data and scored on the test data. 

Several models are scored using above schema , one with the best score i.e in our case the least error is chosen.

**I have used K-fold cross validation method to score the models with variable hyper-paramters**.

The k-fold cross validation method is implemented in file cross-validation.py

```python
# Tuning the hyper-paramters -> degree of polynomial
# fig,axes = plt.subplots(nrows=)
cvr_data=[]
indices= np.arange(0,X.shape[0])
np.random.shuffle(indices)
posssible_degree = np.arange(0, 14, 1)
for k in [4,5,10]:
    cvr = []
    for degree in posssible_degree:
        cvr.append(kfold_cross_validation(X[indices], Y[indices], degree, k,method='piv'))
    cvr_data.append((k,cvr))
```

```python
train_err=[]
for degree in posssible_degree:
    model=polyfit(degree,method='piv')
    model.fit(X,Y)
    train_err.append(model.train_error)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(18,5))
axes = np.array(axes)

for cvr,ax in zip(cvr_data,np.reshape(axes,-1)):
    ax.plot(posssible_degree, np.log(cvr[1]),'-o')
    ax.plot(posssible_degree,np.log(train_err),color='red')
    ax.legend(['CVR K-fold ='+str(cvr[0]),'Training error'])
    ax.set_ylabel('Log(error)')

    ax.set_xlabel('degree')
plt.tight_layout()    
```

```python
# Perform shuffled CVR
shuffled_cvr_data=[]
posssible_degree = np.arange(0, 14, 1)
for k in [4,5,10]:
    cvr = []
    for degree in posssible_degree:
        cvr.append(shuffled_cvr(X,Y,degree=degree,K=k,method='piv',n_permutations=10))
    shuffled_cvr_data.append((k,cvr))
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(18,5))
axes = np.array(axes)

for cvr,ax in zip(shuffled_cvr_data,np.reshape(axes,-1)):
    ax.plot(posssible_degree, np.log(cvr[1]),'-o')
    ax.plot(posssible_degree,np.log(train_err),'-o',color='red')
    ax.legend(['Shuffled CVR K-fold ='+str(cvr[0]),'Training error'])
    ax.set_ylabel('Log(error)')
   
    ax.set_xlabel('degree')
```

#### Observations:
1. The points 20 are too low in number thus we *donot* get the an expected $U$ shaped curved depicting decrese in $(bias)^2$ and increase in $variance$ ,giving an sweet-spot in between.
2. The error is contributed from following
    1. test variance (= variance due to test sample size) : Since the test points are too low in our model thus the test error itsef have a high variance.
    2. **Model instability** variance due to training sample size
    3. **testing bias** 

Thus we see at lower degree of polynomial the cross validation is dominated by bias in model ,but due to so low count of data points the bias does-not dominate that well over the data , hence the varince starts dominating soon.

At $degree =9$ there is sudden dip in the bias , hence we get a see spot where the total bias and variance sum is minimized. 
For $degree\geq 10$ , the varince shoot up

### Hence polynomial of $degree =9$ fits our data well (low train and test error)

### region ($degree<9$ high training error -> Underfitting)
### region ($degree \geq 10 $ high testing error -> overfitting)

## Estimate the noise variance

# Introduce the regularization (lmda)

Introducing paramters $\lambda$ to prevent overfitting case , i.e prevent our $w$ to take abrupt values 
$$E(w,\lambda)_trainingError = \sum{(h(x_i)-t_i)^2} + \lambda (||w||)^2$$

We intriduce a regularized error to our model

## Cross Validation (regulaized model)

Regularization involves adding of regularized error to our data set ,which reduces the hypothesis space ,and forces the parameter $w$ to take values near $0$. Since we derives in class that paramter $\lambda = \frac{\alpha}{\beta}$. Where the $\frac{1}{\alpha}$ is the variance of our hypothesis space under the assumptions of gaussian distribution with 0 mean. Thus increasing values of $\lambda$ would force the model paramter to take values close to 0 and hence we reduce the $variance$ of our model's test error.


**Since we are given only 20 points** , thus the model has a very high $test-error_{variance}$ , we reduce it by regularizing our hypothesis space.

Interested values of $\lambda = [10^{-1},10^{-2},10^{-3},10^{-4}]$

```python
# Tuning the hyper-paramters -> degree of polynomial
# fig,axes = plt.subplots(nrows=)
cvr_data=[]

indices= np.arange(0,X.shape[0])
np.random.shuffle(indices)
posssible_degree = np.arange(3, 15, 1)
k=10
lmda=[1e-1,1e-3,5e-3,1e-2]
for lm in lmda:
    cvr = []
    for degree in posssible_degree:
        # cvr.append(shuffled_cvr(X,Y,degree=degree,K=k,method='piv',n_permutations=20,lmda=lm))
        cvr.append(kfold_cross_validation(X,Y,degree=degree,method='piv',lmda=lm,max_iter=1000))
    cvr_data.append((k,cvr))
```

```python
train_err=[]
for lm in lmda:
    tvr=[]
    for degree in posssible_degree:
        model=polyfit(degree,method='piv',lmda=lm)
        model.fit(X,Y)
        tvr.append(model.train_error)
    train_err.append((lm,tvr))
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(20,5))
axes = np.array(axes)

for cvr,ax,lm,tvr in zip(cvr_data,np.reshape(axes,-1),lmda,train_err):
    ax.plot(posssible_degree, np.log(cvr[1]),'-o')
    ax.plot(posssible_degree,np.log(tvr[1]),'-x')
    ax.legend(['CVR K-fold ='+str(cvr[0])+' Lambda ='+str(lm),'Training error'])
    ax.set_ylabel('Log(error)')
   
    ax.set_xlabel('degree')
```

### ***Since the $degree=11$*** now fit's our model very nicely , we would pick model with $M=11$

### Tuning of the hyperparameter $\lambda$ for $degree=11$

We plot the $error$ vs $-\ln{\lambda}$  ,for both training and test error on our model with $degree =11$

```python
crs,degree=[],11
train_err=[]
for lmd in lmda:
    # compute the test_error  using k-fold cross validation
    crs.append(kfold_cross_validation(X,Y,degree=degree,method='piv',lmda=lmd))
    model=polyfit(degree=degree,method='piv',batch_size=1,lmda=lmd)
    model.fit(X,Y)
    train_err.append(model.score(X,Y))
    
```

```python
plt.figure(figsize=(15,5))
plt.plot(-np.log10(lmda),np.log(train_err),'-o')
plt.plot(-np.log10(lmda),np.log(crs),'-x')
plt.title('$degree=11$',fontsize=24),plt.xlabel('-($\log_{10}{\lambda}$)',fontsize=18),plt.ylabel('$\log_{10}{E_{LSE}}$',fontsize=18)
plt.legend(['train_error','test_error'])
plt.show()
```

# Hence for $\lambda= 10^{-2.6}$ and $degree=11$ , we have have the sweet-spot for our model

```python
# Tuning the hyper-paramters -> degree of polynomial
# fig,axes = plt.subplots(nrows=)
cvr_data=[]

indices= np.arange(0,X.shape[0])
posssible_degree = np.arange(3, 18, 1)
k=5
lmda=[1e-1,1e-3,5e-3,1e-2]
for lm in lmda:
    cvr = []
    for degree in posssible_degree:
        # cvr.append(shuffled_cvr(X,Y,degree=degree,K=k,method='piv',n_permutations=20,lmda=lm))
        cvr.append(kfold_cross_validation(X,Y,degree=degree,method='piv',lmda=lm,max_iter=1000))
    cvr_data.append((k,cvr))
```

```python
train_err=[]
for lm in lmda:
    tvr=[]
    for degree in posssible_degree:
        model=polyfit(degree,method='piv',lmda=lm)
        model.fit(X,Y)
        tvr.append(model.train_error)
    train_err.append((lm,tvr))
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(20,5))
axes = np.array(axes)

for cvr,ax,lm,tvr in zip(cvr_data,np.reshape(axes,-1),lmda,train_err):
    ax.plot(posssible_degree[:18], (cvr[1])[:18],'-o')
    ax.plot(posssible_degree[:18],np.log((tvr[1]))[:18],'-x')
    ax.legend(['CVR K-fold ='+str(cvr[0])+' Lambda ='+str(lm),'Training error'])
    ax.set_ylabel('(error)')
   
    ax.set_xlabel('degree')
plt.tight_layout()
plt.show()
```

### ***Since the $degree=12$*** now fit's our model very nicely , we would pick model with $M=11$

### Tuning of the hyperparameter $\lambda$ for $degree=11$

We plot the $error$ vs $-\ln{\lambda}$  ,for both training and test error on our model with $degree =11$

```python
crs,degree=[],11
train_err=[]
for lmd in lmda:
    # compute the test_error  using k-fold cross validation
    crs.append(kfold_cross_validation(X,Y,degree=degree,method='piv',lmda=lmd))
    model=polyfit(degree=degree,method='piv',batch_size=1,lmda=lmd)
    model.fit(X,Y)
    train_err.append(model.score(X,Y))
    
```

```python
plt.figure(figsize=(15,5))
plt.plot(-np.log10(lmda),np.log(train_err),'-o')
plt.plot(-np.log10(lmda),np.log(crs),'-x')
plt.title('$degree=11$',fontsize=24),plt.xlabel('-($\log_{10}{\lambda}$)',fontsize=18),plt.ylabel('$\log_{10}{E_{LSE}}$',fontsize=18)
plt.legend(['train_error','test_error'])
plt.show()
```

# Hence for $\lambda= 10^{-2.6}$ and $degree=11$ , we have have the sweet-spot for our model
