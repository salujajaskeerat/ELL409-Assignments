```python
%reload_ext autoreload
%autoreload 2
import numpy as np
import matplotlib.pyplot as plt
from polyfit import *
import pandas as pd
from cross_validation import *
from datetime import datetime 
```

```python
df =pd.read_csv('./train.csv',parse_dates=True,index_col='id')
df=df.sort_values(by='id')
df.reset_index(inplace=True)
df.head(3)
```

```python
df['year'] = [d.year -2000 for d in df.id]
df['month'] = [d.month-1 for d in df.id]
df.head()
```

```python
x_y = np.array(df['year']) 
x_m = np.array(df['month']) 
T = np.array(df['value'])
x_t = np.array(((x_y)*12 +x_m),dtype=int)
df['time']= x_t
```

```python
df.head()
```

## Multi linear regression models

```python
months=[[] for i in range(0,12)] # month[i] represent the index of month_i in x3 labels of ith months
for i in range (0,len(x_t)):
    months[(x_t[i]%12)].append(i)   

months
```

```python

plt.figure(figsize=(15,5))
for i in range(0,12):
    
    plt.plot(x_t[np.array(months[i])],T[np.array(months[i])])
plt.show()
```

# Lets train the models for each month , since each month has a cyclic trend over the period of time

models=[]
c
for i i

```python
models,degree=[],5
for i in range(0,12):
    #train the ith month model by fitting a polyfit 
    X_train = x_t[months[i]]
    t = T[months[i]]

    #train a polynomial of degree 2 into our model
    model=polyfit(degree=degree,lmda=1e-3,max_iter=10000,method='piv')
    model.fit(X_train,t)
    models.append(model)
    
```

```python
plt.figure(figsize=(15,5))
for i in range (0,12,2):
    
    plt.plot(models[i].X,models[i].T)
    plt.plot(models[i].X,models[i].pred(models[i].X))
plt.show()
```

```python
model_error = [model.train_error for model in models]
rmse = np.sqrt(np.sum(model_error))/len(x_t)
print(model_error)
rmse
```

# Second Attempt
 Train all the models separately find best fit for each of the model

```python
X=np.reshape(np.array(x_t),(-1,1))
T=np.reshape(np.array(T),(-1,1))
T[months[0]],X[months[0]]
```

## January

```python
# Tuning the hyper-paramters -> degree of polynomial
# fig,axes = plt.subplots(nrows=)
cvr_data=[]

posssible_degree = np.arange(0, 9, 1)
for k in [4,5,10]:
    cvr = []
    for degree in posssible_degree:
#         print(T[months[0]])
        cvr.append(kfold_cross_validation(X[np.array(months[0])], T[np.array(months[0])], degree, k,method='mgrad'))
    cvr_data.append((k,cvr))
```

```python
train_err=[]
for degree in posssible_degree:
    model=polyfit(degree,method='mgrad',steps=1e-3)
    model.fit(X[months[0]],T[months[0]])
    train_err.append(model.train_error)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(18,5))
axes = np.array(axes)

for cvr,ax in zip(cvr_data,np.reshape(axes,-1)):
    ax.plot(posssible_degree, np.log(cvr[1]),'-o')
    ax.plot(posssible_degree,np.log(train_err),color='red')
    ax.legend(['CVR K-fold ='+str(cvr[0]),'Training error'])
    ax.set_ylabel('Log(error)')
   
    ax.set_xlabel('degree')
```

#### for january degree=5 is good
